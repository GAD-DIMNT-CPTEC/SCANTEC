# SCANTEC - Sistema Comunitário de Avaliação de modelos Numéricos de Tempo e Clima

_Versão: 2.0.0b1_

O SCANTEC (Sistema de Avaliação de modelos Numéricos de Tempo e Clima), é um software robusto escrito em linguagem Fortan para a avaliação objetiva de modelos numéricos. A sua aplicação fornece as estatísticas básicas para a qualificação dos modelos através da sua habilidade de previsão. Neste notebook, será feita uma introdução à utiliação do SCANTEC, desde a sua obtenção a partir do repositório principal, até a sua compilação e uso.

Além de apresentar o SCANTEC para a comunidade de usuário do software, este tutorial foi também preparado para introduzir o uso do ambiente do Jupyter a partir da máquina `ilopolis`, acessível a partir do host http://ilopolis.cptec.inpe.br/.

Siga o tutorial na sequência a seguir para aprender a utilizar o SCANTEC. Os links nesta página serão abertos em uma nova aba no seu navegador.

# 1. Obtendo o código

O SCANTEC compreende um conjunto de rotinas escritas em linguagen Fortran que facilitam a tarefa de ler arquivos descritores do GrADS (arquivos `*.ctl`), dados armazenados no formato _GRIdded Binary_ (GRIB) e dados binários (padrão Fortran). Além disso, o SCANTEC é capaz de realizar a interpolação bilinear das grades dos dados a fim de que os cálculos das estatísticas básicas, i.e., do Viés, da Raiz do Erro Quadrático Médio e da Correlação de Anomalias, sejam feitos sempre com base em única grade de referência.

Para cumprir com os objetivos deste tutorial, o SCANTEC poderá ser executado na máquina `itapemirim`. Para ter acesso a esta máquina, siga as instruções a seguir:

1. Obtenha uma cópia local do SCANTEC a partir do link https://projetos.cptec.inpe.br/attachments/download/8937/SCANTEC.2.0.0b1.tar. 

2. Com o pacote salvo no disco, transfira-o para a máquina `itapemirim` e aloque-o em um local adequado (e.g., em `/scripts/grupo/usuario`):

```bash
scp SCANTEC.2.0.0b1.tar usuario@itapemirim.cptec.inpe.br:/scripts/grupo/usuario
```

3. Na interface principal do Jupyter notebook, no canto superior direito, clique em "New", depois em "Terminal" no final da lista;

![abrindo_o_terminal.png](img/abrindo_o_terminal.png)

4. Em uma nova aba do seu navegador, uma seção de terminal estará disponível dentro do Jupyter. Realize a conexão na máquina Itepemirim:

```bash
ssh usuario@itapemirim.cptec.inpe.br -XC
```

5. Escolha um local adequado para baixar o código do SCANTEC, por exemplo, em `/scripts/grupo/usuario`;

```bash
cd /scripts/grupo/usuario
```

6. Descompacte o pacote com o comando:
```bash
tar -xvf SCANTEC.2.0.0b1.tar
```
Com isso, tem-se à disposição uma cópia do SCANTEC para um primeiro teste. Lembre-se que, dependendo do tamanho do seu experimento, pode ser mais adequado executar o SCANTEC em uma máquina mais robusta.

# 2. Compilação

Para compilar o SCANTEC é necessário apenas uma compilador Fortran e um pré-compilador C. A suíte GCC (>=5.X) é suficiente.

Com uma cópia local do código do SCANTEC, a compilação deve ser feita da seguinte forma:

1. Entrar no diretório raiz da sua cópia local do SCANTEC;

```bash
cd SCANTEC.2.0.0b1
```

2. Executar o comando a seguir, selecionando a opção `2 - GNU - Load GNU Gfortran Environment to LINUX and compile`:

```bash
./install
```

![resumo_compilacao.png](img/resumo_compilacao.png)

Aguarde a compilação das bibliotecas do SCANTEC e das rotinas que compõe o kernel do sistema. Se a compilação ocorrer de forma correta, executável `scantec.x` deverá ser criado no diretório `bin`.

# 3. Uso

O SCANTEC é um software projetado para ser utilizado através da linha de comando. Para isto, o usuário deverá estar familiarizado com os seus arquivos de configuração que, na versão atual do software, incluem o namelist principal `scantec.vars` e um arquivo de definições específico para cada modelo cuja extensão é `.model`.

Para encontrar e revisar estes arquivos, o usuário deverá acessar o diretório tables a partir da raiz de instalação do SCANTEC:

![dir_tables.png](img/dir_tables.png)

Observe que, na versão atual, o SCANTEC distribui os seguintes arquivos de definição:

* `BAM_TQ0299L064_18levs.model`: definições do modelo BAM, TQ0299L064 com 18 níveis pós-processados;
* `ETA_ams_05km_22levs.model`: definições do modelo ETA, 5 Km e 22 níveis pós-processados;
* `BRAMS_5km_19levs.model`: definições do modelo BRAMS, 5 Km e 19 níveis pós-processados;
* `BAM_TQ0299L064_28levs.model`: definições do modelo BAM, TQ0299L064 com 28 níveis pós-processados;
* `AGCM_TQ0062L028_50YR_CLIMATOLOGY_18levs.model`: definições da climatologis do modelo AGMC, TQ0062L028 com 18 níveis pós-processados.

Caso seja necessário, o usuário poderá utilizar um destes arquivos como exemplo para criar um novo arquivo de definições de acordo com a grade do seu modelo.

No exemplo, observe que o SCANTEC será utilizado para avaliar as previsões de 24, 48 e 72 horas do modelo BAM (TQ0299L064, com 18 níveis pós-processados), entre 2014080500 e 2014080600. Nesse caso, o intervalo entre as análises é de 12 horas e o intervalo entre as previsões é de 24 horas. As avaliações são feitas contra a própria análise do modelo (`GPOSNMC%y4%m2%d2%h2%y4%m2%d2%h2P.icn.TQ0299L064.ctl`) sobre a região definida pelos limites `49,875S-11.375N` e `82,625W-35,375W` (América do Sul). Observe também que a mensagem `Uso climatologia: 0` indica que no lugar da climatologia, está sendo utilizada a própria análise também.

Para padronizar e flexibilizar a plotagem dos resultados das tabelas, foi preparado um módulo escrito em linguagem Python capaz de ler estas tabelas e plotar as informações de formas diferentes. Este módulo, denominado SCANPLOT, contém funções que permitem transformar as tabelas com as estatísticas do SCANTEC em DataFrames do Pandas (uma estrutura de dados da linguagem Python), facilitanto a exploração e a plotagem das informações. Na presente versão do SCANPLOT, não há funções para a plotagem dos campos espaciais, as quais deverão ser disponibilizadas em versões futuras do módulo. Por enquanto, os usuários são orientados a utilizar o software _Grid Analysis and Display System_ (GrADS) para a plotagem dos campos espaciais.

A apresentação do SCANPLOT e um tutorial para a sua utilização, podem ser acessados em [Plotando Gráficos com o SCANPLOT](Tutorial_SCANPLOT.ipynb) (este link será aberto em um nova página).
